```python
import pandas as pd
import requests
import json
import os
import numpy as np
from tqdm import tqdm
```

```python
file_list = os.listdir('preprocess_data/')
```

```python
total_list = []
code_list = []
for code in tqdm(file_list):
    code_list.append(code)
    df_list = []
    for file_name in os.listdir('preprocess_data/{}'.format(code)):
        df_list.append(pd.read_csv('preprocess_data/{}/{}'.format(code,file_name)))
    df = pd.concat(df_list,axis=0)
    df = df.replace('--',np.nan)
    df = df.replace('</tr>',np.nan)
    df_names = df['报告日期']
    del df['报告日期']
    df.astype('float32')
    df['2019-09-30'] = df.apply(lambda x:x['2019-09-30'] if x['2019-09-30'] != np.nan  else np.mean(x),axis=1)
    df_new = df['2019-09-30'].values.T
    total_list.append(df_new)
```

```python
df = pd.DataFrame(total_list,columns=df_names)
df['code'] = code_list
```
