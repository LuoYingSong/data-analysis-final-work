```python
tele = pd.read_csv('tele.csv')
trans = pd.read_csv('trans.csv')
medical = pd.read_csv('medical.csv')
car = pd.read_csv('car.csv')
total_df_list = [tele,trans,medical,car]
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
def show_reason(predict,label):
    for i in range(np.unique(label).shape[0]):
        tp = ((predict == i) & (test_y == i)).sum()
        tn = ((predict != i) & (test_y != i)).sum()
        fn = ((predict != i) & (test_y == i)).sum()
        fp = ((predict == i) & (test_y != i)).sum()
        print('{}tp:{} tn:{},fn:{},fp:{}'.format(i,tp,tn,fn,fp))
    print(precision_score(predict,label,average='macro'))
    print(recall_score(predict,label,average='macro'))
    print(f1_score(predict,label,average='macro'))
```
